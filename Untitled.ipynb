```python
locations = [(2, 3), (3, 4), (1, 2), (2, 2), (3, 3), (4, 2), (2, 1),
            (9, 6), (10, 5), (9, 7), (8, 6), (10, 7), (11, 4), (9, 5),
            (13, 2), (14, 3), (12, 3), (13, 1), (15, 2), (14, 1)]
```

Você pode estar vendo 2 ou 3 grupos e assim seria fácil de resolver problemas como este, bastaria apenas plotar um grafico, mas imagine em dados com várias dimensões isso seria impossível para uma pessoa visualizar. Por isso usamos modelos como o KMeans para trabalhos como este.

```python
# Oparâmetro n_clusters representa em quantos grupos você deseja agrupar os dados,
# já o random_state serve para definir a semente do gerador de números aleatórios, garantindo a reprodutibilidade dos resultados.
kmeans = KMeans(n_clusters = 3, random_state = 42)

kmeans.fit(locations)
```

```python
# Obter os rótulos dos clusters para cada ponto
labels = kmeans.labels_
```

```python
plt.figure(figsize = (8, 6))

plt.scatter(X, y, label = "Funcionários")

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'blue', marker = 'X')

plt.title("Agrupamento de funcionários em 3 clusters")
plt.show()
```
